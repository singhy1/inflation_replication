```python
data = pd.read_csv(f"{data_dir}/mongey_data/mongey_data.csv")
```

```python
# Define a function to convert Quarter to datetime
def quarter_to_datetime(quarter):
    year = int(quarter)
    fraction = quarter - year
    if fraction == 0.00:
        month = 3
    elif fraction == 0.25:
        month = 6
    elif fraction == 0.50:
        month = 9
    elif fraction == 0.75:
        month = 12
    return pd.Timestamp(year=year, month=month, day=1)

# Apply the conversion to the Quarter column
data['date'] = data['Quarter'].apply(quarter_to_datetime)
```

```python
data = data.rename(columns = {'u_c':'u_rate', 'uhatf_c':'u_rate_job_finding', 'uhats_c':'u_rate_job_destruction'})
```

```python
data
```

```python
data['u_rate'] = data['u_rate']*100 
data['u_rate_job_finding'] = data['u_rate_job_finding']*100
data['u_rate_job_destruction'] = data['u_rate_job_destruction']*100
```

```python
data_pre = data[data['date'] <= '2019-12-01']
```

```python
data_pre
```

```python
# Create the plot 
plt.figure(figsize=(12, 6)) 
    
# Plot both lines
plt.plot(data_pre['date'], data_pre['u_rate'], label='Actual', color='blue')
plt.plot(data_pre['date'], data_pre['u_rate_job_finding'], label='Contribution of Job-Finding', color='red')
    
# Customize the plot
plt.title('Contribution of Job-Finding on Unemployment')
plt.ylabel('Unemployment (pp)')
plt.legend()
# Remove the top and right spines
ax = plt.gca()  # Get current axis
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.savefig(f"{output_dir}/figures/u_rate_comparison_job_finding_pre.pdf", bbox_inches='tight')
plt.show()
```
