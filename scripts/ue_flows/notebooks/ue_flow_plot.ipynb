```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Specify directories 
output_dir = "C:/Users/singhy/Desktop/Chicago/cps_data/inflation/output"
 
```

```python
data = pd.read_csv(f"{output_dir}/data/ue_flows.csv")
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data_16_24 = data[data['date'] >= '2016-01-01']
```

```python
data_16_24 
```

```python
# Create the plot 
plt.figure(figsize=(12, 6)) 
    
# Plot both lines
plt.plot(data_16_24['date'], data_16_24['job_finding_rate'])
# Customize the plot
plt.title('UE')
#plt.ylim(.15,.35)
plt.legend()
plt.savefig(f"{output_dir}/figures/ue_16_24.pdf", bbox_inches='tight')
plt.show()
```

```python
data.set_index('date', inplace = True)
data_quarterly = data.resample('Q').mean()

data_quarterly.reset_index(inplace=True)
```

```python
# Create the plot 
plt.figure(figsize=(12, 6)) 
    
# Plot both lines
plt.plot(data_quarterly['date'], data_quarterly['job_finding_rate'])
# Customize the plot
plt.title('UE')
plt.ylim(.15,.35)
plt.legend()
plt.savefig(f"{output_dir}/figures/ue_long.pdf", bbox_inches='tight')
plt.show()
```

```python
data_quarterly_16 = data_quarterly[data_quarterly['date'] >= '2016-01-01']
```

```python
# Create the plot 
plt.figure(figsize=(12, 6)) 
    
# Plot both lines
plt.plot(data_quarterly_16['date'], data_quarterly_16['job_finding_rate'])
# Customize the plot
plt.title('UE')
plt.ylim(.15,.35)
plt.legend()
plt.savefig(f"{output_dir}/figures/ue_quarterly.pdf", bbox_inches='tight')
plt.show()
```
