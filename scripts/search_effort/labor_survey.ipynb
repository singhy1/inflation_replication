```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data_dir = "C:/Users/singhy/Desktop/Chicago/cps_data/inflation/raw_data"
df = pd.read_excel(f"{data_dir}/sce-labor-microdata-public.xlsx", sheet_name="Data", skiprows = [0])
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m')
```

```python
df
```

```python
select = ['userid', 'date', 'l1m', 'l1y', 'l3', 'js5', 'js6', 'js7', 'js9', 'nl1']
data = df[select]
```

```python
data.rename(columns={
    'userid': 'user_id',
    'date': 'survey_date',
    'js5' : 'job_search_unemp', 
    'js6': 'job_search_emp',
    'js7': 'search_length',
    'js9': 'search_hours',
    'nl1': 'job_offers_rec'
}, inplace=True)
```

```python
# Create emp_stat: 1 if l1y is not NaN, 0 if it is NaN
data['emp_stat'] = data['l1y'].notna().astype(int)
```

```python
# js5 : have you done any job search - unemployed 
# js6 : have you done any job search - employed 
# js7 : length of search 
# js9 : total hours spent searching 
# nl1 : recieved job offer in the last 4 months 
# nl3 : accept/reject the job offer 
```

```python
data.columns
```

```python
keep = ['user_id', 'survey_date', 'job_search_unemp',
       'job_search_emp', 'search_length', 'search_hours', 'job_offers_rec',
       'emp_stat']
data = data[keep]
```

```python
# Clean up the job search indicators 
```

```python
data[data['job_search_unemp'] == 1]
```

```python
data = data[data['survey_date'] > '2015-11-01']
data = data.dropna(subset=["job_search_emp"])
data = data[(data['search_hours'].isna()) | (data['search_hours'] <= 168)]
```

```python
data
```

```python
# Update employment Status 
data['job_search_ind'] = np.where(data['job_search_ind'].isin([1, 2]), 1, 0)
```

```python
data.head()
```

```python
data['search_share'] = data.groupby('survey_date')['job_search_ind'].transform('mean')
```

```python
plot_data = data[['survey_date', 'search_share']].drop_duplicates()

# Step 3: Plot the data
plt.figure(figsize=(10, 6))
plt.plot(plot_data['survey_date'], plot_data['search_share'], marker='o', linestyle='-')
plt.title('Share of Respondents Searching for a Job')
plt.grid(True)
plt.ylim(.15, .3)
plt.show()
```

```python
plot_data = data[data['job_search_ind'] == 1]
```

```python
plot_data = plot_data[plot_data['search_length'] < 200] 
```

```python
plot_data
```

```python
plot_data
```

```python
# Step 1: Calculate the average search_length and search_hours by survey_date
avg_search_length = plot_data.groupby('survey_date')['search_length'].mean()
avg_search_hours = plot_data.groupby('survey_date')['search_hours'].mean()
avg_job_offers = plot_data.groupby('survey_date')['job_offers_rec'].mean()

# Step 2: Create the plots

# Plot 1: Average Search Length
plt.figure(figsize=(10, 6))
plt.plot(avg_search_length.index, avg_search_length.values, marker='o', linestyle='-', color='blue')
plt.title('Average Search Length')
plt.xlabel('Survey Date')
plt.ylabel('Weeks')
plt.grid(True)
plt.ylim(5, 25)
plt.show()

# Plot 2: Average Search Hours
plt.figure(figsize=(10, 6))
plt.plot(avg_search_hours.index, avg_search_hours.values, marker='o', linestyle='-', color='green')
plt.title('Average Search Hours Last Week')
plt.xlabel('Survey Date')
plt.ylabel('Hours')
plt.grid(True)
plt.ylim(2, 6)
plt.show()

# Plot 3: Average Job Offers recieved 
plt.figure(figsize=(10, 6))
plt.plot(avg_job_offers.index, avg_job_offers.values, marker='o', linestyle='-', color='red')
plt.title('Average Job Offers Recieved in the last 4 months')
plt.xlabel('Survey Date')
plt.ylabel("Offers")
plt.grid(True)
plt.ylim(0,1.5)
plt.show()
```
