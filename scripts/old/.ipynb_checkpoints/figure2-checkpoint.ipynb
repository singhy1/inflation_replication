```python
filepath = "C:/Users/singhy/Dropbox (BFI)/Labor_Market_PT/output/cps/figure2_data.dta"
fig_filepath = "C:/Users/singhy/Desktop/Chicago/cps_data/inflation/output/figures"
```

```python
df = pd.read_stata(filepath)
```

```python
df
```

```python
inf = df[df["period"] == "Jul21-Dec22"]
select = ['p90_rwgt_inf', 'p80_rwgt_inf', 'p70_rwgt_inf',
       'p60_rwgt_inf', 'p50_rwgt_inf', 'p40_rwgt_inf', 'p30_rwgt_inf',
       'p20_rwgt_inf', 'p10_rwgt_inf']
inf = inf[select]
```

```python
# Reshape the DataFrame from wide to long format
inf_long = inf.melt(var_name='percentile', value_name='value')

# Extract percentile numbers
inf_long['percentile'] = inf_long['percentile'].str.extract('(\d+)').astype(int)

# Sort by percentile
inf_long = inf_long.sort_values('percentile')

# Create the plot
plt.figure(figsize=(15, 8))
plt.plot(inf_long['percentile'], inf_long['value'], marker='o', label = "July21-Dec22")
plt.ylim(-5,2)
plt.axhline(y=0,color='black', linewidth=1)
plt.xlabel('Percentile')
plt.ylabel('% Annualized change in real hourly wages')
plt.legend()
plt.title("Real Wage Growth Relative to pretrend")
```

```python
post = df[df["period"] == "Jul21-May24"]
select = ['p90_rwgt_post', 'p80_rwgt_post', 'p70_rwgt_post',
       'p60_rwgt_post', 'p50_rwgt_post', 'p40_rwgt_post', 'p30_rwgt_post',
       'p20_rwgt_post', 'p10_rwgt_post']
post = post[select]
```
